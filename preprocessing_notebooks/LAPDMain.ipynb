```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/531-Datasets/LAPD/LAPD_Main.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Get all column names
column_names = df.columns

# Display the column names
print(column_names)
```

```python
# Specify the columns you want to keep
columns_to_keep = ['Weapon Used Cd', 'DATE OCC', 'Weapon Desc', 'Crm Cd', 'Crm Cd Desc', 'Vict Age', 'Vict Descent', 'Vict Sex', 'Premis Desc', 'AREA NAME']

# Create a new DataFrame with only the specified columns
df_filtered = df[columns_to_keep].copy()  # Make a copy to avoid the SettingWithCopyWarning

# Convert the 'DATE OCC' column to datetime format using .loc
df_filtered['DATE OCC'] = pd.to_datetime(df_filtered['DATE OCC'])

# Extract the year and month into new columns using .loc
df_filtered['Year'] = df_filtered['DATE OCC'].dt.year
df_filtered['Month'] = df_filtered['DATE OCC'].dt.month

# Drop rows with any null values
df_cleaned = df_filtered.dropna()

total_rows = df_cleaned.shape[0]
print("Total number of rows After cleaning:", total_rows)

df_cleaned.head()
```

```python
# Get unique value counts for each column
for column in df_cleaned.columns:
    unique_counts = df_cleaned[column].value_counts()
    print(f"Column: {column}\n{unique_counts}\n")
```

```python
# Specify the target number of records
total_records = 12000

# Create an empty DataFrame to store the sampled records
df_sampled = pd.DataFrame()

# Iterate over unique 'AREA NAME' values
for area_name in df_cleaned['AREA NAME'].unique():
    # Select records for the current 'AREA NAME'
    area_records = df_cleaned[df_cleaned['AREA NAME'] == area_name]

    # Determine the number of records to sample for the current 'AREA NAME'
    records_to_sample = min(total_records // len(df_cleaned['AREA NAME'].unique()), len(area_records))

    # Sample records and concatenate to the new DataFrame
    df_sampled = pd.concat([df_sampled, area_records.sample(records_to_sample, random_state=42)])


```

```python
# Get unique value counts for each column
for column in df_sampled.columns:
    unique_counts = df_sampled[column].value_counts()
    print(f"Column: {column}\n{unique_counts}\n")
```

```python
# Display the total number of unique values in each column
for column in df_sampled.columns:
    unique_count = df_sampled[column].nunique()
    print(f"Column: {column}, Total unique values: {unique_count}")
```

```python
# Display all unique values in the 'Premis Desc' column
unique_values = df_sampled['Premis Desc'].unique()

# Print the unique values
print("Unique values in 'Premis Desc' column:")
for value in unique_values:
    print(value)
```
